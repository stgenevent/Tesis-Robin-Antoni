```python
import pandas as pd
```

```python
# Membuat dataset
data = {
    'Bulan': ['September', 'Oktober', 'November', 'Desember', 'Januari', 'Februari'],
    'Mahasiswa_Aktif': [1180, 1195, 1210, 1225, 1240, 1258],
    'Permintaan_Layanan': [8420, 8650, 9340, 10980, 10120, 9730],
    'Persentase_Layanan_Digital': [62, 64, 66, 71, 69, 68]
}

df = pd.DataFrame(data)
data
```

```python
# Statistik deskriptif
deskriptif = df[['Mahasiswa_Aktif', 'Permintaan_Layanan', 'Persentase_Layanan_Digital']].describe()
print(deskriptif)
```

```python
# Korelasi antar variabel
korelasi = df[['Mahasiswa_Aktif',
               'Permintaan_Layanan',
               'Persentase_Layanan_Digital']].corr()

print(korelasi)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Variabel X dan Y
X = df[['Mahasiswa_Aktif', 'Persentase_Layanan_Digital']]
y = df['Permintaan_Layanan']
X
y
```

```python
# Model regresi
model = LinearRegression()
model.fit(X, y)

# Koefisien
print("Intercept:", model.intercept_)
print("Koefisien:", model.coef_)
print("R2:", model.score(X, y))
```

```python
import matplotlib.pyplot as plt
```

```python
df['Prediksi_Regresi'] = model.predict(X)

plt.figure(figsize=(10, 5))

plt.plot(df['Bulan'], df['Permintaan_Layanan'], marker='o', label='Data Aktual')
plt.plot(df['Bulan'], df['Prediksi_Regresi'], marker='o', label='Prediksi Regresi Linier')

plt.title('Perbandingan Permintaan Layanan Aktual dan Prediksi Regresi Linier')
plt.xlabel('Bulan')
plt.ylabel('Permintaan Layanan')
plt.xticks(rotation=30)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(df['Mahasiswa_Aktif'], df['Permintaan_Layanan'])

# Memberikan label nama bulan pada setiap titik
for i in range(len(df)):
    plt.annotate(
        df['Bulan'][i],
        (df['Mahasiswa_Aktif'][i], df['Permintaan_Layanan'][i]),
        textcoords="offset points",
        xytext=(5, 5)
    )

plt.title('Hubungan Jumlah Mahasiswa Aktif dengan Permintaan Layanan')
plt.xlabel('Mahasiswa Aktif')
plt.ylabel('Permintaan Layanan')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(df['Persentase_Layanan_Digital'], df['Permintaan_Layanan'])

# Memberikan label nama bulan pada setiap titik
for i in range(len(df)):
    plt.annotate(
        df['Bulan'][i],
        (df['Persentase_Layanan_Digital'][i], df['Permintaan_Layanan'][i]),
        textcoords="offset points",
        xytext=(5, 5)
    )

plt.title('Hubungan Persentase Layanan Digital dengan Permintaan Layanan')
plt.xlabel('Persentase Layanan Digital (%)')
plt.ylabel('Permintaan Layanan')
plt.grid(True)
plt.tight_layout()

plt.show()
```
